```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import numpy as np

from IPython import display
display.clear_output()
```

```python
iris = load_iris()
predictor = iris.data
classes = iris.target
```

```python
classe_dummy = np_utils.to_categorical(classes)
classe_dummy
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(predictor, classe_dummy, test_size = 0.3, random_state = 0)
```

```python
model = Sequential()
model.add(Dense(units = 5, input_dim = 4))
model.add(Dense(units = 4))
model.add(Dense(units = 3, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.fit(X_train, Y_train, epochs = 1000, validation_data = (X_test, Y_test))
```

```python
predict = model.predict(X_test)
predict = (predict > 0.5)
predict
```

```python
y_test_matrix = [np.argmax(x) for x in Y_test]
predict_matrix = [np.argmax(x) for x in predict]
confusion = confusion_matrix(y_test_matrix, predict_matrix)
confusion
```

```python
from keras.layers import Dropout
import matplotlib.pyplot as plt
from keras.datasets import mnist
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
X_train = X_train.reshape((len(X_train), np.prod(X_train.shape[1:])))
X_test = X_test.reshape((len(X_test), np.prod(X_test.shape[1:])))
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train /= 255
X_test /= 255
```

```python
Y_train = np_utils.to_categorical(Y_train, 10)
Y_test = np_utils.to_categorical(Y_test, 10)
```

```python
model = Sequential()
model.add(Dense(units = 64, activation = 'relu', input_dim = 784))
model.add(Dropout(0.2))
model.add(Dense(units = 64, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 64, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 10, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.fit(X_train, Y_train, epochs = 20, validation_data = (X_test, Y_test))
```

```python
predict = model.predict(X_test)
predict
```

```python
Y_test_matrix = [np.argmax(x) for x in Y_test]
predict_matrix = [np.argmax(x) for x in predict]
confusion = confusion_matrix(Y_test_matrix, predict_matrix)
confusion
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
import pandas as pd
```

```python
data = pd.read_csv('Credit2.csv', sep = ';')
```

```python
X = data.iloc[:, 1:10].values
Y = data.iloc[:, 10].values
```

```python
label = LabelEncoder()
X[:, 0] = label.fit_transform(X[:, 0])
```

```python
onehotencoder = make_column_transformer((OneHotEncoder(categories='auto', sparse = False), [1]), remainder = 'passthrough')
X = onehotencoder.fit_transform(X)
```

```python
labelencoder_Y = LabelEncoder()
Y = labelencoder_Y.fit_transform(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
classifier = Sequential()
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 13))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
predict = classifier.predict(X_test)
predict
```
