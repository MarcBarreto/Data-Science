```python
#pip install nltk
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string

from IPython import display
display.clear_output()
```

```python
corpus = PlaintextCorpusReader('Arquivos', '.*', encoding='ISO-8859-1')
```

```python
file = corpus.fileids()
file[0]
#file[0:10]
```

```python
text = corpus.raw()
```

```python
words = corpus.words()
```

```python
stopword = stopwords.words('english')
#stopword
```

```python
cloud = WordCloud(background_color= 'white', colormap = ListedColormap(['orange', 'green', 'red', 'magenta']), stopwords = stopword, max_words = 180)
cloud.generate(text)
plt.imshow(cloud)
```

```python
words_without_punctuation = [p for p in words if p not in string.punctuation]
```

```python
freq = nltk.FreqDist(words_without_punctuation)
freq
```

```python
most_commons = freq.most_common(50)
most_commons
```
