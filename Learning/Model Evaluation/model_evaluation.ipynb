```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import *

from IPython import display
display.clear_output()
```

```python
data = pd.read_csv('credit3.csv', sep=';')
data.head()
```

```python
data.shape
```

```python
data.groupby('class').size()
```

```python
predictor = data.iloc[:, 1:10].values
classes = data.iloc[:, 10].values
```

```python
label1 = LabelEncoder()
predictor[:, 0] = label1.fit_transform(predictor[:, 0])
label2 = LabelEncoder()
predictor[:, 1] = label2.fit_transform(predictor[:, 1])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(predictor, classes, test_size = 0.3, random_state = 123)
```

```python
forest = RandomForestClassifier()
forest.fit(X_train, Y_train)
```

```python
forest.estimators_
```

```python
predict = forest.predict(X_test)
conf_matrix = confusion_matrix(Y_test, predict)
conf_matrix
```

```python
plot_confusion_matrix(forest, X_test, Y_test)
```

```python
accuracy = accuracy_score(Y_test, predict)
accuracy
```

```python
failure = 1 - accuracy
recall = recall_score(Y_test, predict)
recall
```

```python
precision = precision_score(Y_test, predict)
precision
```

```python
f1 = f1_score(Y_test, predict)
f1
```

```python
auc = roc_auc_score(Y_test, predict)
auc
```

```python
log = log_loss(Y_test, predict)
log
```

```python
from sklearn.svm import SVC
svm = SVC(decision_function_shape='ovo')
svm.fit(X_test, Y_test)
```

```python
predict_svm = svm.predict(X_test)
confusion_svm = confusion_matrix(Y_test, predict_svm)
plot_confusion_matrix(svm, X_test, Y_test)
```

```python
accuracy_svm = accuracy_score(Y_test, predict_svm)
accuracy_svm
failure_svm = 1 - accuracy_svm
```

```python
f1_svm = f1_score(Y_test, predict_svm)
f1_svm
```

```python
auc_svm = roc_auc_score(Y_test, predict_svm)
auc_svm
```
