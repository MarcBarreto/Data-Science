```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score

from IPython import display
display.clear_output()
```

```python
dataset = pd.read_csv('Credit.csv')
dataset.head()
```

```python
classes = dataset.iloc[:, 20].values
variables = dataset.iloc[:, 0:20].values
```

```python
label_encoder1 = LabelEncoder()
variables[:, 0] = label_encoder1.fit_transform(variables[:, 0])
label_encoder2 = LabelEncoder()
variables[:, 2] = label_encoder2.fit_transform(variables[:, 2])
label_encoder3 = LabelEncoder()
variables[:, 3] = label_encoder3.fit_transform(variables[:, 3])
label_encoder4 = LabelEncoder()
variables[:, 5] = label_encoder4.fit_transform(variables[:, 5])
label_encoder5 = LabelEncoder()
variables[:, 6] = label_encoder5.fit_transform(variables[:, 6])
label_encoder6 = LabelEncoder()
variables[:, 8] = label_encoder6.fit_transform(variables[:, 8])
label_encoder7 = LabelEncoder()
variables[:, 9] = label_encoder7.fit_transform(variables[:, 9])
label_encoder8 = LabelEncoder()
variables[:, 11] = label_encoder8.fit_transform(variables[:, 11])
label_encoder9 = LabelEncoder()
variables[:, 13] = label_encoder9.fit_transform(variables[:, 13])
label_encoder10 = LabelEncoder()
variables[:, 14] = label_encoder10.fit_transform(variables[:, 14])
label_encoder11 = LabelEncoder()
variables[:, 16] = label_encoder11.fit_transform(variables[:, 16])
label_encoder12 = LabelEncoder()
variables[:, 18] = label_encoder12.fit_transform(variables[:, 18])
label_encoder13 = LabelEncoder()
variables[:, 19] = label_encoder13.fit_transform(variables[:, 19])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(variables, classes, test_size = 0.3, random_state = 0)
X_train
```

# **Naive Bayes**

```python
naive_bayes = GaussianNB()
naive_bayes.fit(X_train, Y_train)
```

```python
predict = naive_bayes.predict(X_test)
predict
```

```python
confusionMatrix = confusion_matrix(Y_test, predict)
confusionMatrix
```

```python
accuracy = accuracy_score(Y_test, predict)
accuracy
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(iris.data)
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
kmeans_confusion = confusion_matrix(iris.target, kmeans_predict)
kmeans_confusion
```

```python
plt.scatter(iris.data[kmeans_predict == 0, 0], iris.data[kmeans_predict == 0, 1], c = 'green', label = 'Setosa')
plt.scatter(iris.data[kmeans_predict == 1, 0], iris.data[kmeans_predict == 1, 1], c = 'red', label = 'Versicolor')
plt.scatter(iris.data[kmeans_predict == 2, 0], iris.data[kmeans_predict == 2, 1], c = 'blue', label = 'Virginica')
plt.legend()
```

# **KMedoids**

```python
#pip install pyclustering
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.cluster import cluster_visualizer
```

```python
kmedoids = kmedoids(iris.data[:, 0:2], [3, 12, 20])
kmedoids.get_medoids()
```

```python
kmedoids.process()
kmedoids_predict = kmedoids.get_clusters()
kmedoids_predict
```

```python
kmedoids_visualizer = cluster_visualizer()
kmedoids_visualizer.append_cluster(kmedoids_predict, iris.data[:, 0:2])
medoids = kmedoids.get_medoids()
kmedoids_visualizer.append_cluster(medoids, data = iris.data[:, 0:2], marker = '*', markersize = 20)
kmedoids_visualizer.show()
```

```python
predict_list = []
real_list = []
for i in range(len(kmedoids_predict)):
     for j in range(len(kmedoids_predict[i])):
        predict_list.append(i)
        real_list.append(iris.target[kmedoids_predict[i][j]])
```

```python
predict_list = np.array(predict_list)
real_list = np.array(real_list)
kmedoids_confusion = confusion_matrix(real_list, predict_list)
kmedoids_confusion
```

# **Association Rules**

```python
#pip install apyori
from apyori import apriori
```

```python
transactions = pd.read_csv('transacoes.txt', header = None)
```

```python
transactions_list = []
for i in range(0, len(transactions)):
    transactions_list.append([str(transactions.values[i, j]) for j in range(0, 3)])
transactions_list
```

```python
rules = apriori(transactions_list, min_support = 0.5, min_confidence = 0.5, min_lenght = 2)
```

```python
result = list(rules)
result
```

```python
result2 = [list(x) for x in result]
result2
```

```python
result3 = []
for j in range(0, 7):
        result3.append([list(x) for x in result2[j][2]])
result3
```
