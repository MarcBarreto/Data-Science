```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from pyclustertend import hopkins, vat, ivat, assess_tendency_by_metric, assess_tendency_by_mean_metric_score
```

```python
iris = load_iris()
kMeans = KMeans(n_clusters = 3)
cluster = kMeans.fit_predict(iris.data)
```

```python
plt.title('Iris')
plt.scatter(iris.data[:, 0], iris.data[:, 1], c = cluster, cmap='rainbow')
plt.show()
```

```python
cluster1 = scale(iris.data)
cluster2 = np.random.rand(150, 4)
```

```python
inertia = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(iris.data)
    inertia.append(kmeans.inertia_)
plt.plot(range(1, 8), inertia)
plt.title('Elbow')
plt.show()
```
