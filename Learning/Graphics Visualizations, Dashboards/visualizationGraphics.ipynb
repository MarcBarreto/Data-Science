```python
from pandas import *
from statistics import *
from seaborn import *
```

```python
data = read_csv("dados.csv", sep = ";")
data.head()
```

```python
#Histograma do Empenho sem linha de densidade
distplot(data.VALOREMPENHO, hist = True, kde = False)
#Linha de densidade sem histograma
distplot(data.VALOREMPENHO, hist = False, kde = True)
#Histograma com linha de densidade
distplot(data.VALOREMPENHO).set_title("Interest")
```

```python
# GDP Graph Bar
data.PIB.plot.bar()
```

```python
#Pie Graph
data.VALOREMPENHO.plot.pie()
```

```python
# GDP Boxplot Graph
boxplot(dados.PIB).set_title("GDP")
```

```python
# Boxplot Graph
boxplot(data = data)
```

```python
# Grafico de disperso
scatterplot(data.VALOREMPENHO, data.PIB, marker = "+", color = "blue")
```

```python
from matplotlib.pyplot import *
#GDP Line Graph
plot(data.PIB)
#Interest Line Graph
plot(data.Empenho)
boxplot(data.PIB, showfliers = False, notch = True, patch_artist = True).set_title("PIB")
```
