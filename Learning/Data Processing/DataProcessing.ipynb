```python
from pandas import *
from statistics import *
from seaborn import *
```

```python
dados = read_csv("tempo.csv", sep = ";")
dados.head()
dados.shape
dados.isnull()
```

```python
dados.groupby(["Aparencia"]).size()
dados.loc[(dados["Aparencia"] != "sol") & (dados["Aparencia"] != "chuva") & (dados["Aparencia"] != "nublado"), "Aparencia"] = "sol"
dados.groupby(["Aparencia"]).size()
```

```python
dados["Temperatura"].describe()
dados.loc[(dados["Temperatura"] < -130) | (dados["Temperatura"] > 130), "Temperatura"] = median(dados["Temperatura"])
dados["Temperatura"].describe()
```

```python
dados["Umidade"].describe()
mediana = median(dados["Umidade"])
dados["Umidade"].fillna(mediana, inplace = True)
dados.loc[(dados["Umidade"] < 0) | (dados["Umidade"] > 100), "Umidade"] = mediana
dados["Umidade"].describe()
```

```python
dados.groupby(dados["Vento"]).size()
dados["Vento"].fillna("FALSO", inplace = True)
dados.isnull()
```
