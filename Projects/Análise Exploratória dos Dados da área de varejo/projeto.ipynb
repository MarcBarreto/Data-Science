## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
total_vendas = df_dsa[['Data_Pedido', 'Valor_Venda']].groupby(['Data_Pedido'])['Valor_Venda'].sum()
plt.figure(figsize =(20, 6))
total_vendas.plot(x = 'Data_Pedido', y = 'Valor_Venda')
plt.title('Total de Vendas por Data do Pedido')
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.bar(total_vendas.index, total_vendas.values)
plt.xlabel('Data do Pedido')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por Data do Pedido')
plt.show()
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
vendas_desconto15 = df_dsa.query('Valor_Venda > 1000')['Valor_Venda'].count()
vendas_desconto15
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
df_desconto = df_dsa.copy()
df_desconto['Desconto'] = np.where(df_desconto['Valor_Venda'] > 1000, 0.15, 0.10)
df_desconto['Valor_desconto'] = df_desconto['Valor_Venda'] - (df_desconto['Valor_Venda'] * df_desconto['Desconto'])
vendas_antes_desconto = df_desconto.loc[df_desconto['Desconto'] == 0.15, 'Valor_Venda']
vendas_depois_desconto = df_desconto.loc[df_desconto['Desconto'] == 0.15, 'Valor_desconto']
print(f'Valor medio das vendas antes do desconto: {vendas_antes_desconto.mean()}')
print(f'Valor medio das vendas depois do desconto: {vendas_depois_desconto.mean()}')
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
df_dsa['Ano'] = df_dsa['Data_Pedido'].str.split('/').str[2]
df_dsa['Mes'] = df_dsa['Data_Pedido'].str.split('/').str[1]
media_vendas_sam = df_dsa.groupby(['Ano', 'Mes', 'Segmento'])['Valor_Venda'].mean()
segmento_index = media_vendas_sam.index.get_level_values(2)
ano_index = media_vendas_sam.index.get_level_values(0)
mes_index = media_vendas_sam.index.get_level_values(1)
```

```python
plt.figure(figsize=(12,6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                   y = media_vendas_sam.values,
                   x = mes_index,
                   hue = segmento_index,
                   col = ano_index,
                   col_wrap = 4)
plt.show()
```

## Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python
vendas_sub_categoria = df_dsa.groupby(['Categoria',
                             'SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',
                                                                                   ascending = False).head(12)
```

```python
vendas_sub_categoria = vendas_sub_categoria[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
```

```python
vendas_sub_categoria_cat = vendas_sub_categoria.groupby('Categoria').sum(numeric_only = True).reset_index()
```

```python
# Listas de cores para categorias
cores_categorias = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
```

```python
# Listas de cores para subcategorias
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
# Plot

# Tamanho da figura
fig, ax = plt.subplots(figsize = (18,12))

# Gráfico das categorias
p1 = ax.pie(vendas_sub_categoria_cat['Valor_Venda'], 
            radius = 1,
            labels = vendas_sub_categoria_cat['Categoria'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = cores_categorias)

# Gráfico das subcategorias
p2 = ax.pie(vendas_sub_categoria['Valor_Venda'],
            radius = 0.9,
            labels = vendas_sub_categoria['SubCategoria'],
            autopct=lambda pct: f'${int(pct / 100 * sum(vendas_sub_categoria["Valor_Venda"]))}',
            colors = cores_subcategorias, 
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'), 
            pctdistance = 0.53,
            rotatelabels = True)

# Limpa o centro do círculo
centre_circle = plt.Circle((0, 0), 0.6, fc = 'white')

# Labels e anotações
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(vendas_sub_categoria['Valor_Venda']))), xy = (-0.2, 0))
plt.title('Total de Vendas Por Categoria e Top 12 SubCategorias')
plt.show()
```
